```
#import pandas and attempt to read csv
import pandas as pd
df = pd.read_csv('selfie_analysis/selfie_dataset.csv')
```

```
#view first 10 records
df.head(10)
```

```
#view last 10 records
df.tail(10)
```

```
is_female = df.groupby('is_female').count()['image_name']
baby = df.groupby('baby').count()['image_name']
child = df.groupby('child').count()['image_name']
teenager = df.groupby('teenager').count()['image_name']
youth = df.groupby('youth').count()['image_name']
middle_age = df.groupby('middle_age').count()['image_name']
senior = df.groupby('senior').count()['image_name']

d = pd.DataFrame({'is_female': is_female, 'baby': baby, 'child': child, 
                 'teenager': teenager, 'youth': youth, 
                 'middle_age': middle_age, 'senior': senior})
d.T.plot.bar(rot=0)
```
